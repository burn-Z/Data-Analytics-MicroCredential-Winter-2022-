# Task 1 : Environment Set up

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

:Import all required libraries that are needed for data analysis

# Task 2 :  Data Collection

Use nytimes github repository as a data source to collect the covid 19 data in real time and ingest it into colab notebook for data anlysis purpose

```python
df = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv')
df.head(50)
```

# Task 3 : Data Wrangling and EDA (Exploratory Data Analysis)

```python
df_NYC = df[(df['state'] == 'New York') & (df['county'] == 'New York City')].drop(['fips', 'county', 'state'], axis= 1)
df_NYC.info()
```

```python
df_NYC
```

```python
sns.histplot(df_NYC['deaths'], bins= 20)
```

```python
plt.figure(figsize= (12,10))
sns.scatterplot(data= df_NYC, x= 'cases', y= 'deaths')
```

# Task4: Understand NY covid 19 data in last 30 days

```python
df_NYC_30days = df_NYC[df_NYC['date'] >= '2021-12-18']
dates = df_NYC_30days['date']
plt.figure(figsize= (12,10))
sns.lineplot(data= df_NYC_30days, x = 'date', y= 'cases').set_xticklabels(labels= dates,rotation= -90)
plt.figure(figsize= (12,10))
sns.lineplot(data= df_NYC_30days, x = 'date', y= 'deaths').set_xticklabels(labels= dates,rotation= -90)
#.set(title= 'New York City Case count in the last 30 days')
```

```python
df_LA = df[(df['state'] == 'California') & (df['county'] == 'Los Angeles')].drop(['fips', 'county', 'state'], axis= 1)
df_LA_30days = df_LA[df_LA['date'] >= '2021-12-18']

sns.lineplot(data= df_LA_30days, x = 'cases', y= 'deaths')
```

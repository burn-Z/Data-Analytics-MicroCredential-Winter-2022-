Ask the user for a number. Depending on whether the number is even or odd, print out an appropriate message to the user. Hint: how does an even / odd number react differently when divided by 2?


```
user_input = int(input("Enter a number:\t"))
```

```
if(user_input % 2 == 0):
    print(f"{user_input} is an even number")
else:
    print(f"{user_input} is an odd number")
```

# Bonus:

2.	Ask the user for two numbers: one number to check (call it num) and one number to divide by (check). If check divides evenly into num, tell that to the user. If not, print a different appropriate message.


```
# Bonus 1
user_input = int(input("Enter a number:\t"))
```

```
if(user_input % 4 == 0):
    print(f"{user_input} is a multiple of 4")
elif(user_input % 2 == 0):
    print(f"{user_input} is an even number")
else:
    print(f"{user_input} is an odd number")
```

```
# Bonus 2
print("Enter two numbers to check if the first is divisible by the second.\n")
```

```
num = int(input("Enter the first number: "))
check = int(input("Enter the second number: "))
```

```
if(num % check == 0):
    print(f"{num} is evenly divisible by {check}.")
else:
    print(f"{num} is not a multiple of {check}.")
```

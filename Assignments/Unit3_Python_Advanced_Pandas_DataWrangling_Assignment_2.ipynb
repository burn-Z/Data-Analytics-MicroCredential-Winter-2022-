# Median salary between gender and/or rank(title) Etc...


Making a new Data Frame that only has the columns 'sex' and 'salary'

```
new_data = df[['sex','salary']]
```

Finding the Median salary for all males

```
Male_data = new_data[new_data['sex'] == 'Male']
Male_data['salary'].median()
```

Finding the Median salary for all females

```
Female_data = new_data[new_data['sex'] == 'Female']
Female_data['salary'].median()
```

Comparing the Mean salary of Males and Females

```
Male_data['salary'].mean()
```

```
Female_data['salary'].mean()
```

```
# The difference between mean salaries of male and female
Male_data['salary'].mean() - Female_data['salary'].mean()
```

# Create some simple visualizations showing some simple statistics

```
sns.set(rc = {'figure.figsize': (15,10)})
sns.scatterplot(data= df, y= 'salary', x= 'yrs.service', hue= 'sex', palette= 'YlOrBr').set(xlabel= 'Experience')
```

```
sns.scatterplot(data= df, y= 'salary', x= 'yrs.service', hue= 'rank', palette= 'YlOrBr').set(xlabel= 'Experience')
```

```
sns.pairplot(data= df.drop('ID', axis= 1), hue= 'discipline')
```

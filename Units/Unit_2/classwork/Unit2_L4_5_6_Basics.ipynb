```python
tv_shows = ['star trek', 'game of thrones', 'stranger things', 'big bang theory', 'black mirror']

for show in tv_shows[:3]:
    print(show.title())
```

```python
#Conditionals

for show in tv_shows:
    if show == "star trek" or show == 'big bang theory':
        print('I dont watch trash. AKA ' + show)
    elif show == 'stranger things':
        print(show + ' was aight')
    elif show == 'black mirror':
        print(show + " wasn't too bad")
    else:
        print(show + " is not a favorite")
```

```python
#Dictionaries

me = {'first_name':'greatest','last_name':'ever', 'bmi':'thick', 'height':69, 'wealth': 0, 4:20}
print(me['first_name'].title() + me['last_name'].title())
```

```python
del me['bmi']
print(me)

print(me.popitem())
```

```python
for key in me.keys():
    print(key)

for value in me.values():
    print(value)
```

```python
for key, value in me.items():
    print(f"Key: {key}\tValue: {value}\n")
```

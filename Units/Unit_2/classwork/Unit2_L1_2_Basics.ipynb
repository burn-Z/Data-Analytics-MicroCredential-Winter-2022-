```python
# list

movies = ["star wars", "Avengers", "Deadpool", 9, "Kung Fu Panda"]

print(movies)

print(movies[3])

print(movies[-1].upper())

print(movies.index("star wars"))
```

```python
movies[1] = "Life of Pi"
print(movies)

movies.append("Scary Movie")
print(movies)

movies.insert(2, "Of Mice and Men")
print(movies)
```

```python
print(movies.pop())
print(movies)

print(movies.pop(3))
print(movies)

movies.remove("Life of Pi")
print(movies)

del movies[2]
print(movies)
```

```python
"star wars" in movies
```

```python
movies.sort(reverse=True)
print(movies)
```

```python
print(sorted(movies))
print(movies)
```

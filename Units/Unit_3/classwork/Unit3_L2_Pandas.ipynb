```python
# Second dataset

import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/311_Service_Requests_from_2019May.csv')
```

```python
df.info()
```

```python
df.groupby('Borough')['Unique Key'].count()
```

```python
df.groupby('Agency')['Unique Key'].count().sort_values(ascending= False)
```

```python
df.groupby('Agency')['Unique Key'].count().sort_values(ascending= False).plot(kind= 'bar')
```

```python
df_BAC = df[['Borough', 'Agency', 'Complaint Type']]
df_BAC
```

```python
df_queens = df_BAC[df_BAC['Borough'] == 'QUEENS']
df_queens
```

```python
df_BAC_queens_or_manhattan = df_BAC[(df_BAC['Borough'] == 'QUEENS') | (df_BAC['Borough'] == 'MANHATTAN')]
df_BAC_queens_or_manhattan
```

```python
df_BAC_queens_and_dep = df_BAC[(df_BAC['Borough'] == 'QUEENS') & (df_BAC['Agency'] == 'DEP')]
df_BAC_queens_and_dep
```

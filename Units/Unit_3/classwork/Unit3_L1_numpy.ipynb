```python
# shallow and deep copies

NYC_Borough = np.array(['Manhattan','Queens', 'Bronx', 'Staten Island', 'Brooklyn'])
NYC_Borough
```

```python
# shallow copy

view_NYC_Borough = NYC_Borough.view()
view_NYC_Borough
```

```python
view_NYC_Borough[2] = 'wrong'
view_NYC_Borough
```

```python
NYC_Borough
```

```python
# deep copy

copy_NYC_Borough = NYC_Borough.copy()
```

```python
copy_NYC_Borough[3] = 'change'
copy_NYC_Borough
```

```python
NYC_Borough
```

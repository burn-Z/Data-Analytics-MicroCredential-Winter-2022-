```python
# Seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
OG_Pokemon = pd.read_csv('https://raw.githubusercontent.com/negojovanovich3/pythonprojects/main/Pokemon.csv',index_col = 0)
```

```python
OG_Pokemon.info()
```

```python
sns.lmplot(data= OG_Pokemon, x= 'Attack', y= 'Defense').set(xlabel = "Damage", ylabel = "Block", title= "Pokemon: Attack v. Defense")

# zoom in on plot

#plt.xlim(40,100)
#plt.ylim(25,90)
```

```python
# very long runtime
sns.pairplot(data= OG_Pokemon, hue= 'Stage')
```

```python
OG_Pokemon = OG_Pokemon.drop(["Total", 'Legendary'],axis =1 )

# axis 1 is the row

sns.set_style('dark')
sns.boxplot(data= OG_Pokemon)
```

```python
sns.scatterplot(data= OG_Pokemon, x= 'Attack', y= 'Defense', hue= 'Legendary', palette= 'YlOrBr')

#   more palettes
#   Spectral    vlag    icefire     coolwarm    Blues
```

```python
sns.set(rc = {'figure.figsize': (12,10)})
sns.swarmplot(data= OG_Pokemon, x= "Type 1", y= "Attack")
```

```python
#corr = OG_Pokemon.corr()
sns.heatmap(data= OG_Pokemon.corr(), cmap= 'Reds_r' )
```

```python
sns.histplot(data= OG_Pokemon, x= 'Attack', bins = 100)
```

```python
sns.factorplot(data= OG_Pokemon, x = "Type 1", y= 'Attack', kind = 'swarm', col = 'Stage').set_xticklabels(rotation=-45)
```

```python
sns.jointplot(data= OG_Pokemon, x= 'Attack', y= 'Defense', kind= 'hex')     #kind optional
```
